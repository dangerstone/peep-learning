```python
basicbirder, basicbirderinfo = load_function("base_lr0.0005.tar")
basicbirder.eval()
gardenbirder, gardenbirderinfo = load_function("fine_base.tar")
gardenbirder.eval()
print_stats(gardenbirderinfo)
```

```python
import matplotlib.pyplot as plt

epochs = 60

# Combine losses for full training range
trainlosses = basicbirderinfo["train_loss"] + gardenbirderinfo["train_loss"]
vallosses = basicbirderinfo["val_loss"] + gardenbirderinfo["val_loss"]

# Plotting
plt.figure(figsize=(5, 4))
plt.plot(range(1, epochs + 1), trainlosses, label='Training loss')
plt.plot(range(1, epochs + 1), vallosses, label='Validation loss')
plt.axvline(x=50, color='pink', linestyle='--', label='Finetuning begins')

# Annotations and labels
plt.title('Loss per epoch')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
#plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Combine losses for full training range
trainaccs = basicbirderinfo["train_acc"] + gardenbirderinfo["train_acc"]
valaccs = basicbirderinfo["val_acc"] + gardenbirderinfo["val_acc"]

# Plotting
plt.figure(figsize=(5, 4))
plt.plot(range(1, epochs + 1), trainaccs, label='Training accuracy')
plt.plot(range(1, epochs + 1), valaccs, label='Validation accuracy')
plt.axvline(x=50, color='pink', linestyle='--', label='Finetuning begins')

# Annotations and labels
plt.title('Accuracy per epoch')
plt.xlabel('Epoch')
plt.ylim(0.7,1)
plt.ylabel('Accuracy')
plt.legend()
#plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```
